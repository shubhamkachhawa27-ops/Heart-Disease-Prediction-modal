# Heart Disease Prediction Using Machine Learning Classification Models

**Goal:** Develop a supervised machine learning system that predicts whether a patient has heart disease (Target: 0 = No, 1 = Yes) using medical parameters.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import warnings
warnings.filterwarnings('ignore')
```

## 1. Load Dataset


```python
# Ensure dataset is downloaded as heart.csv in data folder
df = pd.read_csv('../data/heart.csv')
display(df.head())

# Check for missing values
print("Missing Values:\n", df.isnull().sum())
df = df.dropna()
```

## 2. Exploratory Data Analysis (EDA)


```python
# Distribution of the target
plt.figure(figsize=(8,5))
sns.countplot(x='target', data=df, palette='viridis', hue='target', legend=False)
plt.title('Heart Disease Frequency (0 = No, 1 = Yes)')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(14,10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

## 3. Data Preprocessing


```python
X = df.drop('target', axis=1)
y = df['target']

# Scaling numeric features
continuous_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
scaler = StandardScaler()

X_scaled = X.copy()
X_scaled[continuous_cols] = scaler.fit_transform(X_scaled[continuous_cols])

# 80/20 train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

## 4. Modeling & Hyperparameter Tuning


```python
# Logistic Regression
lr = LogisticRegression()
lr.fit(X_train, y_train)
print("Logistic Regression Accuracy:", accuracy_score(y_test, lr.predict(X_test)))
```

```python
# Random Forest with GridSearchCV
rf_params = {'n_estimators': [50, 100], 'max_depth': [None, 10]}
rf_grid = GridSearchCV(RandomForestClassifier(random_state=42), rf_params, cv=5)
rf_grid.fit(X_train, y_train)
print("Best RF Params:", rf_grid.best_params_)
print("Random Forest Accuracy:", accuracy_score(y_test, rf_grid.predict(X_test)))
```

## 5. Feature Importance


```python
best_rf = rf_grid.best_estimator_
plt.figure(figsize=(10,6))
sns.barplot(x=best_rf.feature_importances_, y=X.columns, palette="Blues_r", hue=X.columns, legend=False)
plt.title('Feature Importances from Random Forest')
plt.show()
```
